```python
# number of active member by gender
data.groupby(['gender'])['active_member'].value_counts()
```

```python
active_males = data[(data['active_member'] == 1) & (data['gender'] == 'Male')].shape[0]
print(active_males)
```

```python
#estimated salary less than 50000
data2=data[data['estimated_salary']< 50000]
print(data2)
```

```python
# how many males estimated salary greater than 45000
data.groupby([data['estimated_salary']>45000])['gender'].value_counts()
```

```python
num_males = data[(data['estimated_salary'] > 45000) & (data['gender'] == 'Male')].shape[0]
print(num_males)
```

```python
# top 5 customer id whose tenure is more than 5 years
data.groupby([data['tenure']>5])['customer_id'].head(3)
```

```python
tenure = data[(data['tenure'] > 5) & (data['customer_id'])].head(5)
print(tenure)
```

```python
# customer id whose credit score is greater than 750
credit = data[(data['credit_score'] > 750) & (data['customer_id'])].shape[0]
print(credit)
```
