```python
from matplotlib import pyplot as plt
```

```python
x = [6, 3, 6, 8, 7, 9, 5, 0]
y = [10, 5, 8, 5, 2, 8, 6, 5]
plt.plot(x, y)
plt.show()
```

```python
x = [3, 6, 1, 4, 7, 9]
y = [10, 5, 7, 4, 3, 1]
plt.bar(x, y)
plt.show()
```

```python
y = [10, 3, 8, 7, 1]
plt.hist(y)
plt.show()
```

```python
x = [5, 2, 9, 4, 2, 4, 2, 3, 6, 2, 5]
y = [10, 5, 8, 4, 2, 3, 7, 2, 7, 8, 0]
plt.scatter(x, y)
plt.show()
```

```python
labels = ['Academy', 'IT Services', 'BPO Services']
sizes = [55, 35, 30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart Example')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1, y2 = [10, 20, 10, 15, 30], [15, 25, 10, 20, 25]

plt.fill_between(x, y1, y2, color='blue', alpha=0.4, label='Area 1-2')
plt.plot(x, y1, label='Line 1', marker='*')
plt.plot(x, y2, label='Line 2', marker='o')

plt.xlabel('X-axis'), plt.ylabel('Y-axis'), plt.title('Area Chart Example')

plt.legend(), plt.show()
```

```python
fig = plt.figure(figsize=(3, 3), facecolor='green', layout='constrained')
fig.suptitle('Figure')
ax = fig.add_subplot()
ax.set_title('Axes', loc='left', fontstyle='oblique', fontsize='medium')
```

```python
fig,a =  plt.subplots(2,2)
import numpy as np
x = np.arange(2,6)
a[0][0].plot(x,x*x)
a[0][0].set_title('square')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('square root')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exp')
a[1][1].plot(x,np.log10(x))
a[1][1].set_title('log')
plt.show()
```
