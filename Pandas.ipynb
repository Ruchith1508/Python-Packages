# Replacing Missing Values

```python
df=pd.read_csv(r'Titanic-Pandas.csv',usecols=['Age','Survived','Fare'])
```

```python
df.head(12)
```

```python
mean_value=df['Age'].mean()
print(mean_value)
```

```python
data2=df.fillna(mean_value)
```

```python
data2.head(12)
```

```python
median_value=df['Age'].median()
print(median_value)
```

```python
data3=df.fillna(median_value)
```

```python
data3.head(12)
```

# Aggregating Data with groupby

```python
df=pd.read_csv(r'Titanic-Pandas.csv',usecols=['Age','Survived','Fare','Cabin'])
```

```python
mean_value=df['Age'].mean()
print(mean_value)
```

```python
df.fillna(mean_value).head(6)
```

```python
df.groupby('Cabin').mean().head()
```
